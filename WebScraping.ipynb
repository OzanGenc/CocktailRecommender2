```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from random import randint
from time import sleep
import pickle
from google.colab import files
```

## Getting the List of Cocktails

```python
drink_categories = ['creamy-cocktails', 'highball-cocktails', 'hot-cocktails', 'fruit-herb-cocktails', 
                    'low-proof-cocktails', 'savory-cocktails', 'simple-cocktails', 'sour-cocktails', 
                    'sparkling-cocktails', 'spirit-forward-cocktails', 'tropical-cocktails']

drink_link_list = []

for category in drink_categories:

  url_drinks = 'https://cocktailpartyapp.com/drink-category/' + category + '/'

  r_drinks = requests.get(url_drinks)

  # Extracts the response as html: html_doc
  html_doc = r_drinks.text

  # Create a BeautifulSoup object from the HTML: soup
  soup_drinks = BeautifulSoup(html_doc)

  # Prettify the BeautifulSoup object: pretty_soup
  #pretty_soup_drinks = soup_drinks.prettify()


  i = 0 #indice of a drink

  var = soup_drinks.findAll('div', attrs={'class':'archive-content'})[0].findAll('a', attrs={'class':'thumb'})
  #var = soup_drinks.findAll('div', attrs={'class':'archive-content'})[0].findAll('a', attrs={'class':'thumb'})[i].attrs['href']
  

  for i in range(len(var)):

    drink_link_list.append(var[i].attrs['href'])

  # Sleep a random number of seconds (between 1 and 5)
  sleep(randint(10,20))
```

# Scraping Multiple Cocktails

```python
data_list = []
k = 1


for url in drink_link_list:


  try:
    # Specify url: url
    #url = 'https://cocktailpartyapp.com/drinks/alpine-toddy/'

    # Package the request, send the request and catch the response: r
    r = requests.get(url)

    # Extracts the response as html: html_doc
    html_doc = r.text

    # Create a BeautifulSoup object from the HTML: soup
    soup = BeautifulSoup(html_doc)

    # Prettify the BeautifulSoup object: pretty_soup
    pretty_soup = soup.prettify()

    # Print the response

    pd_title = soup.title.text.split(' |')[0]
    pd_recipe_description = soup.findAll('div', attrs={'class':'recipe-description'})[0].text.replace("\n", " ")
    pd_instructions = soup.findAll('div', attrs={'class':'recipe-instructions'})[0].text.replace("\n", " ")

    ingredients = soup.findAll('span', attrs={'class':'ingredient'})

    ingredient_str = str()

    for i in range(len(ingredients)):

      ingredient_str = ingredients[i].text.replace("\n", "").replace("\r", "").replace("\t", "") + ", " + ingredient_str 

    pd_ingredient = ingredient_str 

    #data_list = []

    info = 'Ingredients: ' + pd_ingredient + '. Instructions: ' + pd_instructions + 'Description: ' + pd_recipe_description

    data_dict = {'name': pd_title , 'ingredients': pd_ingredient, 'instructions': pd_instructions, 'description': pd_recipe_description}

    data_list.append(data_dict)

    

  except:
    print("Data couldn't be retrieved from the url: {}".format(url))

  print(str(k) + ' / 1297')
  k += 1
  sleep(randint(10,20))

  if k % 100 == 0:

    df = pd.DataFrame(data_list)

    df.to_pickle("./cocktails_df.pkl") 

    files.download("cocktails_df.pkl")

    sleep(randint(100,200))




df = pd.DataFrame(data_list)

df.to_pickle("./cocktails_df.pkl") 

files.download("cocktails_df.pkl")
```
